```python
import numpy as np
import matplotlib.pyplot as plt

# Set the parameters for the sine wave
amplitude = 1.0  # Amplitude of the wave
frequency = 1.0  # Frequency of the wave in Hz
duration = 10.0   # Duration of the wave in seconds
sampling_rate = 100  # Number of samples per second (standard for audio)
desired_sampling_rate=30
# Generate the time axis
t = np.linspace(0, duration, int(sampling_rate * duration), endpoint=False)

# Generate the sine wave
sine_wave = amplitude * np.sin(2 * np.pi * frequency * t+np.pi/7)
print(len(sine_wave))
downsampled_sine_wave = [sine_wave[j] for j in range(0, len(sine_wave), desired_sampling_rate)]
new_t =np.linspace(0, duration,len(downsampled_sine_wave), endpoint=False)
print(len(downsampled_sine_wave))
# Plot the sine wave
plt.plot(new_t,downsampled_sine_wave)
plt.plot(t, sine_wave)

plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Sine Wave')
plt.show()
```
